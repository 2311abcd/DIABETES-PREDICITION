```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/diabetes (1).csv')
```

```python
# seperating the data and labels
x  =df.drop(columns='Outcome',axis=1)
y = df['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
# DATA STANDARDISATION
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
standardized_data = scaler.transform(x)
```

```python
print(standardized_data)
```

```python
x = standardized_data
y=df['Outcome']
```

```python
print(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier_pred = classifier.predict(x_train)
accuracy = accuracy_score(classifier_pred,y_train)
```

```python
print(f'ACCURACY SCORE OF TRAINING DATA :{accuracy*100:.2f}%')
```

```python
classifier_pred_test = classifier.predict(x_test)
accuracy_test = accuracy_score(classifier_pred_test,y_test)
```

```python
print(f'ACCURACY SCORE OF TESTING DATA :{accuracy_test*100:.2f}%')
```

```python
import numpy as np

# Use correct number of features (8)
input_data = (11,143,94,33,146,36.6,0.254,51)

# Convert to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape for single prediction
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardize
std_data = scaler.fit_transform(input_data_reshaped)
print("Standardized input:", std_data)

# Predict
prediction = classifier.predict(std_data)
#print("Prediction:", prediction)

# Output result
if prediction[0] == 0:
    print("PERSON IS NOT DIABETIC")
else:
    print("PERSON IS DIABETIC")

```
